```python
# data manipulation
import pandas as pd

# data viz
import matplotlib.pyplot as plt
import seaborn as sns

# styling
plt.style.use("ggplot")
```

```python
#Reading archive Iris
planta = pd.read_csv("iris.data")
```

```python
planta.head()
```

```python
#Summary of column information
planta.describe()
```

```python
#Type of data in each column
planta.dtypes
```

```python
#Total number of rows and columns
planta.shape
```

```python
#Renaming columns
planta.columns=['sepala_lenght','sepala_width','petala_lenght','petala_width','especie']
```

```python
planta.head()
```

```python
#Checking if there are any null values in the sepala_length column
planta[planta['sepala_lenght'].isnull()].head()
```

```python
#Counting the total of setosa
planta['especie'].value_counts()
```

```python
#Creating a graph with just one line of code
planta['especie'].value_counts().head(10).plot(kind='bar',figsize=(11,5),grid=False,rot=0,color='orange')
#Making the graphic more pleasing
plt.title('Iris flower dataset')
plt.xlabel('Classification') #naming the x-axis, where the type of iris is located
plt.ylabel('Quantity') #naming the y-axis, where the total rating is
plt.show() #displaying the graph
```
